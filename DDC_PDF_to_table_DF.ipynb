```python
# Step 1: Install tabula-py in the Colab environment
!pip install tabula-py

# Step 2: Download the PDF file
!wget https://datadrivenconstruction.io/wp-content/uploads/2023/07/DDC_Sample_Building_Measure_Savings.pdf

# Step 3: Use tabula-py to extract tables from the PDF

# Import necessary libraries
import tabula
import pandas as pd

# Use tabula-py to extract tables from the downloaded PDF
tables = tabula.read_pdf("DDC_Sample_Building_Measure_Savings.pdf", pages='all', multiple_tables=True)

# Convert the tables to pandas dataframes and display them
for i, table in enumerate(tables):
    print(f"Table {i+1}")
    display(table)
```

```python
import pandas as pd

# Save the tables to an Excel file, with each table on a separate sheet
with pd.ExcelWriter('extracted_tables.xlsx') as writer:
    for i, table in enumerate(tables):
        sheet_name = f"Table {i+1}"
        table.to_excel(writer, sheet_name=sheet_name, index=False)
        print(f"Saved {sheet_name} to Excel")

# At this point, the Excel file 'extracted_tables.xlsx' will be available in your Colab environment.
# You can download it using the Colab UI.
```
